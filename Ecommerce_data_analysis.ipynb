```python
pd.set_option('display.float_format', '{:,.3f}'.format)

monthly_sales = df.groupby(df['Order Date'].dt.month_name().str[:3])['Sales'].sum()
month_order = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
monthly_sales = monthly_sales.reindex(month_order)

monthly_sales.loc['Grand Total'] = monthly_sales.sum()
print(monthly_sales)

```

```python
df.head()
```

#### Trend line

```python
# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='b')

# Add labels and title
plt.title('Monthly Sales Trend', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales (in Millions)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show chart
plt.show()
```

#### Market distribution using Pie chart

```python
market_distribution= df.groupby('Market')['Sales'].sum()
plt.pie(market_distribution, labels = market_distribution.index,
        autopct = '%.0f%%',
        startangle = 90)
plt.title('Market')
plt.axis('equal')
plt.show()
```

#### Sales by Customer Segment

```python

customer_segment= df.groupby("Customer Segment")['Sales'].sum()
# Convert to millions for readability
values = customer_segment.values / 1e6
categories = customer_segment.index

# Plot with thinner bars (width < 0.8 makes bars thinner)
plt.figure(figsize=(8,6))
bars = plt.bar(categories, values, color="royalblue", width=0.4)

# Add labels on top of bars
for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             f"${value:.2f} M", ha='center', va='bottom', fontsize=10, fontweight='bold')

# Titles and labels
plt.ylabel("Sales (in Millions)")
plt.title("Sales by Customer Segment")
plt.show()
```

#### Top 5 Categories by Sales

```python


# Group and get top 5
category_name = df.groupby("Category Name")["Sales"].sum().reset_index()
top5 = category_name.sort_values("Sales", ascending=False).head(5)

# Plot
plt.figure(figsize=(8,3))
ax = sns.barplot(x="Sales", y="Category Name", data=top5, color="royalblue")

plt.title("Top 5 Categories by Sales")
plt.xlabel("Sales")
plt.ylabel("Category Name")

plt.tight_layout()
plt.show()
```

#### Bottom 5 Categories by Sales

```python


# Group and get bottom 5
category_name = df.groupby("Category Name")["Sales"].sum().reset_index()
top5 = category_name.sort_values("Sales", ascending=True).head(5)

# Plot
plt.figure(figsize=(8,3))
ax = sns.barplot(x="Sales", y="Category Name", data=top5, color="royalblue")

plt.title("Bottom 5 Categories by Sales")
plt.xlabel("Sales")
plt.ylabel("Category Name")

plt.tight_layout()
plt.show()
```

###  Top 10 country wise sales

```python

# Group and get top 5
country_name = df.groupby("Order Country")["Sales"].sum().reset_index()
top5 = country_name.sort_values("Sales", ascending=False).head(10)

# Plot
plt.figure(figsize=(8,3))
ax = sns.barplot(x="Sales", y="Order Country", data=top5, color="royalblue")

plt.title("Top 10 Country wise Sale")
plt.xlabel("Sales")
plt.ylabel("Order Country")

plt.tight_layout()
plt.show()
```
